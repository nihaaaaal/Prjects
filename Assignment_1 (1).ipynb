```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('boston.csv')
```

```python
df.head()
```

```python
df["CRIM"].hist(figsize=(10,6))
```

```python
def plotvariale(df, variable):
    plt.figure(figsize=(14,8))
    plt.subplot(1,2,1) 
    df[variable].hist(bins=30)
    
    
    plt.subplot(1,2,2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.show()
```

```python
plotvariale(df, 'CRIM')
```

```python
df["CRIM"] = np.log(df["CRIM"])
```

```python
plotvariale(df, 'CRIM')
```

```python
df_1 = pd.read_csv('boston_1.csv')
```

```python
df_1.head()
```

```python
df_1["CRIM"].hist(figsize=(10,6))
```

```python
plotvariale(df_1, 'CRIM')
```

```python
df_1["CRIM"] = np.log(df_1["CRIM"])
```

```python
plotvariale(df_1, 'CRIM')
```

# Both the datasets are now similarly distributed and identify the same variable once processed. The log transformation is applied to only one column for example.
